```python
from random import randint

default_symboly = ["x", "o"]

max_pocet_tahu = 9
aktualni_tah = 1
pocet_viteznych_symbolu = 3
pouzite_tahy = []

hraci_pole = [["_" for _ in range(3)] for _ in range(3)]
for radek in hraci_pole:
    print(" ".join(radek))

hra_ukoncena = False

while aktualni_tah <= max_pocet_tahu and not hra_ukoncena:
    tah_hrace = [int(input(f"Zvol souřadnici č.1 v rozsahu od 0 do 2: ")), int(input(f"Zvol souřadnici č.2 v rozsahu od 0 do 2: "))]
    if tah_hrace not in pouzite_tahy:
        hraci_pole[tah_hrace[0]][tah_hrace[1]] = default_symboly[0]
        pouzite_tahy.append(tah_hrace)
        aktualni_tah += 1
    else:
        print("Zvol jiné souřadnice")
    for radek in hraci_pole:
        print(" ".join(radek))

    if aktualni_tah > max_pocet_tahu:
        break

    while True:
        tah_pocitace = [randint(0, 2), randint(0, 2)]
        if tah_pocitace not in pouzite_tahy:
            hraci_pole[tah_pocitace[0]][tah_pocitace[1]] = default_symboly[1]
            pouzite_tahy.append(tah_pocitace)
            aktualni_tah += 1
            break
    for radek in hraci_pole:
        print(" ".join(radek))

    vitezny_sloupec = [default_symboly[0]]*pocet_viteznych_symbolu
    for symbol in range(pocet_viteznych_symbolu):
        sloupec = [radek[symbol] for radek in hraci_pole]
        if sloupec == vitezny_sloupec:
            print("Vyhrál jsi!")
            hra_ukoncena = True
            break
        elif [default_symboly[1]]*pocet_viteznych_symbolu == sloupec:
            print("Počítač vyhrál!")
            hra_ukoncena = True
            break
    if hra_ukoncena:
        break
    vitezny_radek = [default_symboly[0]]*pocet_viteznych_symbolu
    for radek in hraci_pole:
        if radek == vitezny_radek:
            print("Vyhrál jsi!")
            hra_ukoncena = True
            break
        elif [default_symboly[1]]*pocet_viteznych_symbolu == radek:
            print("Počítač vyhrál!")
            hra_ukoncena = True
            break
    if hra_ukoncena:
        break
    vitezna_hlavni_diagonala = [default_symboly[0]]*pocet_viteznych_symbolu
    diagonala = [hraci_pole[i][i] for i in range(len(hraci_pole))]
    if diagonala == vitezna_hlavni_diagonala:
        print("Vyhrál jsi!")
        hra_ukoncena = True
        break
    elif [default_symboly[1]]*pocet_viteznych_symbolu == diagonala:
        print("Počítač vyhrál!")
        hra_ukoncena = True
        break
    vitezna_vedlejsi_diagonala = [default_symboly[0]]*pocet_viteznych_symbolu
    vedlejsi_diagonala = [hraci_pole[i][len(hraci_pole)-i-1] for i in range(len(hraci_pole))]
    if vedlejsi_diagonala == vitezna_vedlejsi_diagonala:
        print("Vyhrál jsi!")
        hra_ukoncena = True
        break
    elif [default_symboly[1]]*pocet_viteznych_symbolu == vedlejsi_diagonala:
        print("Počítač vyhrál!")
        hra_ukoncena = True
        break
```
